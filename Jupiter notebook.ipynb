##### Segmenting and Clustering Neighborhoods in Toronto

```python

tabel2 = tablenew.groupby(by=['Postcode','Borough']).agg(lambda x: ','.join(x))
```

```python
tabel2
```

```python
tabel2.reset_index(level=['Postcode','Borough'], inplace=True)
```
